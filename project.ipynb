```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv("netflix_large_sample_500.csv")
df.head()
df.info()
missing_values = df.isnull().sum()

df_clean = df.drop_duplicates()
df_clean = df_clean.reset_index(drop = True)

# movies dtaframe 
df_movies = df_clean[df_clean['Type'] == "Movie"]
print( "Movies are :" , len(df_movies))

# series dataframe 
df_series = df_clean[df_clean['Type'] == 'Series']
print("series are: " , len(df_series))

# top 5 countries with most metflix content 
top_countries = df_clean['Country'].value_counts().head(5)
print("top 5 countries are: " , top_countries)

#most common genre 
common_genre = df_clean['Genre'].value_counts().idxmax()
print("the most common genre are :" , common_genre)

#title having rating TV-MA
tv_ma_count = (df_clean['Rating'] == 'TV-MA').sum()
print("number of tittles having TV-MA rating ia:" , tv_ma_count)

#earliest and latest release year 
earlest_year = df_clean['Release_Year'].min()
print("the earliest release year is :" , earlest_year)

latest_year = df_clean['Release_Year'].max()
print("the latest release year is :" , latest_year)

#number of tittles per year 

title_per_year = df_clean["Title"].value_counts()
print(title_per_year)










```

```python
# bar chart 

import matplotlib.pyplot as plt 
counts = df_clean['Type'].value_counts()
plt.bar(counts.index, counts.values, color=['red', 'blue'])
plt.title("movie vs series ")
plt.xlabel("index")
plt.ylabel("values")
plt.show()
```

```python
# pie chart 

import matplotlib.pyplot as plt 
genre = df_clean['Genre'].value_counts()
plt.pie(genre.values , autopct = '%1.1f%%' , labels = genre.index)
plt.title("genre distribution ")
plt.show()
```

```python
# line plot

import matplotlib.pyplot as plt 
year = df['Release_Year'].value_counts()
plt.bar(year.index , year.values , color = 'red')
plt.grid(color = 'grey' , linestyle = ":")
plt.title("Release per year")
plt.xlabel("year")
plt.ylabel("no of movies")
plt.show()
```

```python
import matplotlib.pyplot as plt 
countries = df_clean['Country'].value_counts().head(5)
plt.bar(countries.index , countries.values , color = 'purple')
plt.title("top 5 country release netflix content ")
plt.grid( color = 'grey' , linestyle = ':')
plt.xlabel('top 5 country')
plt.ylabel("no of content release")
plt.show()
```

```python
ratings = df_clean['Rating'].value_counts()
plt.bar(ratings.index , ratings.values , color = 'red')
plt.title("ratings distribution")
plt.grid(color = 'grey' , linestyle = ':')
plt.xlabel("ratings")
plt.ylabel("no of tittles")     
plt.show()
```
